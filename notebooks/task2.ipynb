```python
data["Message"]=data["Message"].str.upper()
data["Message"]=data["Message"].str.replace(",","")
data["Message"]=data["Message"].str.replace(".","")
data["Message"]=data["Message"].str.replace("?","")
data["Message"]=data["Message"].str.replace("/","")
data["Message"]=data["Message"].str.replace("\\","")
data["Message"]=data["Message"].str.replace("-","")
import re
def remove_emojis(text):
    emoji_pattern = re.compile(
        "[" 
        "\U0001F600-\U0001F64F"  # emoticons
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F680-\U0001F6FF"  # transport & map symbols
        "\U0001F700-\U0001F77F"  # alchemical symbols
        "\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
        "\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
        "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
        "\U0001FA00-\U0001FA6F"  # Chess Symbols
        "\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
        "\U00002702-\U000027B0"  # Dingbats
        "\U000024C2-\U0001F251" 
        "]+", 
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)

# Apply the function to the 'Message' column
data['Message'] = data['Message'].apply(remove_emojis)

# Display the updated DataFrame
data.head()
```

```python
data["Date"]=pd.to_datetime(data["Date"])
```

```python
data['valid_channel_title'] = data['Channel Title'].notna() & (data['Channel Title'].str.len() > 0)
data['valid_channel_username'] = data['Channel Username'].notna() & data['Channel Username'].str.match(r'^\w+$')
data['unique_channel_username'] = ~data['Channel Username'].duplicated()
data['valid_id'] = data['ID'].notna() & data['ID'].apply(lambda x: isinstance(x, int))
data['unique_id'] = ~data['ID'].duplicated()
data['valid_message'] = data['Message'].notna() & (data['Message'].str.len() > 0) & (data['Message'].apply(lambda x: isinstance(x, str)))
data['valid_date'] = data['Date'].notna() & (pd.to_datetime(data['Date'], errors='coerce').notna())
data['valid_date_range'] = data['Date'] <= pd.Timestamp.now(tz='UTC')
data['valid_media_path'] = data['Media Path'].notna() & data['Media Path'].str.len() > 0
data['valid_media_format'] = data['Media Path'].str.endswith(('.jpg', '.png', '.jpeg', '.gif'))
```

```python
import os
import psycopg2
import pandas as pd
from dotenv import load_dotenv
from sqlalchemy import create_engine


# Load environment variables from .env file
load_dotenv()

# Fetch database connection parameters from environment variables
DB_HOST = os.getenv("DB_HOST")
DB_PORT = os.getenv("DB_PORT")
DB_NAME = os.getenv("DB_NAME")
DB_USER = os.getenv("DB_USER")
DB_PASSWORD = os.getenv("DB_PASSWORD")
print(DB_PASSWORD)
```

```python
import psycopg2
import pandas as pd

def insert_data_to_postgres(data, table_name):
    """
    Connects to the PostgreSQL database and inserts data from a DataFrame into the specified table.

    :param data: DataFrame containing the data to insert.
    :param table_name: The name of the table where the data will be inserted.
    """
    try:
        # Establish a connection to the database
        connection = psycopg2.connect(
            host=DB_HOST,
            port=DB_PORT,
            database=DB_NAME,
            user=DB_USER,
            password=DB_PASSWORD
        )

        # Create a cursor object
        cursor = connection.cursor()

        # Prepare the SQL insert statement
        for index, row in data.iterrows():
            # Create a tuple of the row data
            row_data = tuple(row)
            # Generate the insert query dynamically
            insert_query = f"""
            INSERT INTO {table_name} ( channel_title, channel_username,message_Id, message, date, media_path)
            VALUES (%s,%s, %s, %s, %s, %s);
            """
            cursor.execute(insert_query, row_data)

       
        connection.commit()

        print(f"Data inserted successfully into {table_name}!")

    except Exception as e:
        print(f"An error occurred: {e}")

    finally:
        # Close the database connection
        if connection:
            cursor.close()
            connection.close()
            print("Connection closed.")


```
